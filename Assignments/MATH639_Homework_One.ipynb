1) For the vector vield 

$$
{\bf F} = \left( \cos(2t),\sin(t) \right)
$$

find 

$$
\int_{C} {\bf F}\cdot \hat{{\bf n}} ds, ~ \int_{C} {\bf F}\cdot \hat{{\bf t}} ds 
$$

where C is the contour between $(-1,1)$ and $(1,1)$ along the parabola $y=x^{2}$.

**Solution**

So here we have $-1 \leq x \leq 1$ and $0 \leq y \leq x^{2}$. To simplify our computation, we use the parametrization

$$
x(t) = t,
$$

$$
y(t) = t^{2}.
$$

So we can define the vector

$$
{\bf r}(t) = x(t)\hat{{\bf i}}+y(t)\hat{{\bf j}} \longrightarrow {\bf r}(t) = t\hat{{\bf i}}+t^{2}\hat{{\bf j}}.
$$

Let us start by solving 

$$
\int_{C} {\bf F}\cdot \hat{{\bf t}} ds.
$$

We know that

$$
\hat{{\bf t}} = \frac{{\bf r^{\prime}}(t)}{||{\bf r^{\prime}}(t)||}.
$$

So we have

$$
\hat{{\bf t}} = \frac{<1,2t>}{\sqrt{1+4t^{2}}}.
$$

Now we parametrize the arc length $ds$ by doing

$$
\frac{ds}{dt} = \sqrt{\left(\frac{dx}{dt}\right)^{2}+\left(\frac{dy}{dt}\right)^{2}} = \sqrt{1+4t^{2}} = ||{\bf r^{\prime}}(t)||.
$$

Therefore

$$
ds = ||{\bf r^{\prime}}(t)||dt
$$

which implies that

$$
\hat{{\bf t}} ds = \frac{{\bf r^{\prime}}(t)}{||{\bf r^{\prime}}(t)||}||{\bf r^{\prime}}(t)|| = {\bf r^{\prime}}(t)dt,
$$

which makes our job much easier. Now we can compute the scalar product

$$
{\bf F}\cdot \hat{{\bf t}} ds = {\bf F}\cdot{\bf r^{\prime}}(t)dt = \cos(2t)+2t\sin(t).
$$

So, since $-1\leq t \leq 1$, then our integral is

$$
\int_{-1}^{1}(\cos(2t)+2t\sin(t))dt.
$$

Solving a piece at a time

$$
\int_{-1}^{1}\cos(2t)dt = \frac{1}{2}\left[\sin(2t)\right]_{-1}^{1} = \sin(2)
$$

and

$$
\int_{-1}^{1}2t\sin(t)dt = [-2t\cos(t)]_{-1}^{1}+2\int_{-1}^{1}\cos(t) = 4\sin(1)-4\cos(1).
$$

Therefore, in conclusion

$$
\int_{C} {\bf F}\cdot \hat{{\bf t}} ds = \sin(2) + 4\sin(1)-4\cos(1).
$$

Now, for the normal vector, since it is obviously perpendicular to the tangent vector, we have

$$
\hat{\bf n} = \frac{<-2t,1>}{\sqrt{4t^{2}+1}}
$$

So, since $$
ds = {\sqrt{4t^{2}+1}}dt
$$

$$
\int_{C} {\bf F}\cdot \hat{{\bf n}} ds = \int_{-1}^{1}{\bf F}\cdot\frac{<-2t,1>}{\sqrt{4t^{2}+1}}{\sqrt{4t^{2}+1}}dt \longrightarrow \int_{-1}^{1}{\bf F}\cdot{<-2t,1>}dt
$$

So we take the dot product

$$
<\cos(2t)\sin(t)>\cdot{<-2t,1>} = -2t\cos(2t)+\sin(t)
$$

and now we integrate it

$$
\int_{-1}^{1}(-2t\cos(2t)+\sin(t))dt.
$$

Since we know that $\cos(t)=\cos(-t)$, then obviously $\int_{-1}^{1}\sin(t)=0$. Thus we compute by parts

$$
\int_{-1}^{1}-2t\cos(2t)dt = [2t\sin(t)]_{-1}^{1}+[\cos(2t)]_{-1}^{1} = 0.
$$

Therefore, in conclusion

$$
\int_{C} {\bf F}\cdot \hat{{\bf n}} ds = 0
$$

2) Find 

$$
\int\int_{\mathcal{D}} (x^{2} + y^{2})dx dy,
$$

where $\mathcal{D}$ is the domain $1\leq xy \leq 4$, $1\leq y/x \leq 4$.  The transformation $G:\mathbb{R}^{2}\rightarrow \mathbb{R}^{2}$ where

$$
\begin{pmatrix}
x \\ y
\end{pmatrix} = G(u,v) = \begin{pmatrix}
uv^{-1} \\ uv
\end{pmatrix}
$$

makes this problem **much** easier. 

**Solution**
So, the given $f(x,y)$ is a circle of radius $1$ centered at the origin. As the domain $\mathcal{D}$ would take this problem nontrivial then,using the suggested transformation, we go from the domain $\mathcal{D}$ to the region $\mathcal{S}$ defined by

$$
xy=u^{2} \longrightarrow \pm1\leq u \leq \pm2
$$

and

$$
y/x=v^{2} \longrightarrow \pm1\leq v \leq \pm2.
$$

So, basically we have two rectangular subregions of |width|=2 and |height|=1, one in the first quadrant, call it $S_{1}$, and one in the third quadrant, call it $S_{2}$. For simplicity in the computation, we work with the positive subregion and then will multiply our results by $2$ as

$$
\int_{a}^{b}f(x)dx=\int_{-a}^{-b}f(x)dx.
$$

So, first of all we need to compute the Jacobian
\begin{align}
J=  \begin{bmatrix}
    \frac{dx}{du} & \frac{dx}{dv}\\
    \frac{dy}{du} & \frac{dy}{dv}
    \end{bmatrix}.
\end{align}

Thus, since $x=u/v$ and $y=uv$ our Jacobian is

\begin{align}
J=  \begin{bmatrix}
    \frac{1}{v} & -\frac{u}{v^{2}}\\
    v & u
    \end{bmatrix}.
\end{align}

and the determinant is readily found to be

$$
det(J)=2\frac{u}{v}.
$$

So now we solve our integral

$$
\int\int_{\mathcal{D}} (x^{2} + y^{2})dx dy = \int_{1}^{2}\int_{1}^{2}\left(\frac{u^{2}}{v^{2}}+u^{2}v^{2}\right)2\frac{u}{v}dudv \longrightarrow \int_{1}^{2}\int_{1}^{2}2\left(\frac{u^{3}}{v^{3}}+u^{3}v\right)dudv.
$$

We have

\begin{align}
\int_{1}^{2}\left[\frac{u^{4}}{2v^{3}} + \frac{u^{4}v}{2}\right]_{1}^{2}dv\\
= \frac{15}{2}\int_{1}^{2}\frac{1}{v^{3}}dv+\frac{15}{2}\int_{1}^{2}vdv\\
= -\frac{15}{4}\left[\frac{1}{v^{2}}\right]_{1}^{2}+\frac{15}{4}\left[v^{2}\right]_{1}^{2}\\
= \frac{45}{16}+\frac{45}{4}\\
= \frac{225}{16}
\end{align}

As stated before, this is just the area of the subregion $S_{1}$ in the first quadrant, so, to take into account the $S_{3}$ subregion, we multiply our results by $2$. Therefore

\begin{align}
\int\int_{\mathcal{S}}f(x(u,v),y(u,v))dudv\\ =\int\int_{\mathcal{S_{1}}}f(x(u,v),y(u,v))dudv+\int\int_{\mathcal{S_{2}}}f(x(u,v),y(u,v))dudv\\ =2\int\int_{\mathcal{S_{1}}}f(x(u,v),y(u,v))dudv\\
=\frac{225}{8}.
\end{align}

3) Using the spherical coordinate transformation

$$
x = \rho \cos(\theta)\sin(\phi), ~ y = \rho\sin(\theta)\sin(\phi), ~ z = \rho \cos(\phi),
$$

find the Laplacian $\Delta = \partial_{x}^{2} + \partial_{y}^{2} + \partial_{z}^{2}$ in spherical coordinates.  

**Solution**

We first need to find the gradient with respect to spherical coordinates, given by $\triangledown = \frac{\partial}{\partial x} + \frac{\partial}{\partial y} + \frac{\partial}{\partial z}$, where

$$
\frac{\partial}{\partial x} = \frac{\partial}{\partial \rho}\frac{\partial \rho}{\partial x} + \frac{\partial}{\partial \theta}\frac{\partial \theta}{\partial x} + \frac{\partial}{\partial \phi}\frac{\partial \phi}{\partial x}, \\
\frac{\partial}{\partial y} = \frac{\partial}{\partial \rho}\frac{\partial \rho}{\partial y} + \frac{\partial}{\partial \theta}\frac{\partial \theta}{\partial y} + \frac{\partial}{\partial \phi}\frac{\partial \phi}{\partial y}, \\
\frac{\partial}{\partial z} = \frac{\partial}{\partial \rho}\frac{\partial \rho}{\partial z} + \frac{\partial}{\partial \theta}\frac{\partial \theta}{\partial z} + \frac{\partial}{\partial \phi}\frac{\partial \phi}{\partial z}.
$$

From here we use $\rho^{2} = x^{2} + y^{2} + z^{2}$, $\tan{\theta} = \frac{y}{x}$, and $\phi = \arccos{\left(\frac{z}{\sqrt{x^{2} + y^{2} + z^{2}}}\right)}$ to find the partial derivatives needed. 

From $\rho^{2} = x^{2} + y^{2} + z^{2}$ we can compute

$$
2\rho \rho_{x} = 2x \Rightarrow \rho_{x} = \frac{x}{\rho} = \sin{\phi} \cos{\theta}, \\
2\rho \rho_{y} = 2y \Rightarrow \rho_{y} = \frac{y}{\rho} = \sin{\phi} \sin{\theta}, \\
2\rho \rho_{z} = 2z \Rightarrow \rho_{z} = \frac{z}{\rho} = \cos{\phi}.
$$

From $\theta = \arctan{\frac{y}{x}}$ we can compute

$$
\theta_{x} = \frac{-y}{x^{2}}\left( \frac{1}{1+ \frac{y^{2}}{x^{2}}} \right) = \frac{-y}{x^{2}}\left( \frac{1}{1+ \tan^{2}\theta} \right) = \frac{-\sin{\theta}}{\rho \sin{\phi}}, \\
\theta_{y} = \frac{1}{x}\left( \frac{1}{1+ \frac{y^{2}}{x^{2}}} \right) = \frac{1}{x}\left( \frac{1}{1+ \tan^{2}\theta} \right) = \frac{\cos{\theta}}{\rho \sin{\phi}}, \\
\theta_{z} = 0.
$$

From $\phi = \arccos{\left(\frac{z}{\sqrt{x^{2} + y^{2} + z^{2}}}\right)}$ we calculate

$$
\phi_{x} = \frac{xz}{(x^{2} + y^{2} + z^{2})^{3/2}}\left(\frac{1}{\sqrt{1 - \frac{z}{\sqrt{x^{2} + y^{2} + z^{2}}}^{2}}}\right) = \frac{xz}{\rho^{3}}\left(\frac{1}{\sqrt{1 - \cos^{2}\phi}}\right) = \frac{\cos \theta \cos \phi}{\rho}, \\
\phi_{y} = \frac{yz}{(x^{2} + y^{2} + z^{2})^{3/2}}\left(\frac{1}{\sqrt{1 - \frac{z}{\sqrt{x^{2} + y^{2} + z^{2}}}^{2}}}\right) = \frac{yz}{\rho^{3}}\left(\frac{1}{\sqrt{1 - \cos^{2}\phi}}\right) = \frac{\sin \theta \cos \phi}{\rho}, \\
\phi_{z} = \frac{-(1 - \cos^{2}\phi)}{\rho} \left(\frac{1}{\sin \phi}\right) = \frac{-\sin \phi}{\rho}.
$$

Now we get

$$
\frac{\partial}{\partial x} = \sin{\phi} \cos{\theta} \frac{\partial}{\partial \rho} + \frac{-\sin{\theta}}{\rho \sin{\phi}}\frac{\partial}{\partial \theta} + \frac{\cos \theta \cos \phi}{\rho}\frac{\partial}{\partial \phi}, \\
\frac{\partial}{\partial y} = \sin{\phi} \sin{\theta}\frac{\partial}{\partial \rho} + \frac{\cos{\theta}}{\rho \sin{\phi}}\frac{\partial}{\partial \theta} + \frac{\sin \theta \cos \phi}{\rho}\frac{\partial}{\partial \phi}, \\
\frac{\partial}{\partial z} = \cos{\phi}\frac{\partial}{\partial \rho} +  \frac{-\sin \phi}{\rho}\frac{\partial}{\partial \phi}.
$$

Now, finally, we can proceed in finding the Laplacian by first getting its individual components

$$
\frac{\partial^{2}}{\partial x^{2}} = \sin{\phi} \cos{\theta} \left(\frac{\partial \frac{\partial}{\partial x}}{\partial \rho} \right) + \frac{-\sin{\theta}}{\rho \sin{\phi}}\left(\frac{\partial \frac{\partial}{\partial x}}{\partial \theta} \right) + \frac{\cos \theta \cos \phi}{\rho}\left(\frac{\partial \frac{\partial}{\partial x}}{\partial \phi} \right), \\
\frac{\partial^{2}}{\partial y^{2}} = \sin{\phi} \sin{\theta}\left(\frac{\partial \frac{\partial}{\partial y}}{\partial \rho} \right) + \frac{\cos{\theta}}{\rho \sin{\phi}}\left(\frac{\partial \frac{\partial}{\partial y}}{\partial \theta} \right) + \frac{\sin \theta \cos \phi}{\rho}\left(\frac{\partial \frac{\partial}{\partial y}}{\partial \phi} \right), \\
\frac{\partial^{2}}{\partial z^{2}} = \cos{\phi}\left(\frac{\partial \frac{\partial}{\partial z}}{\partial \rho} \right) + \frac{-\sin \phi}{\rho}\left(\frac{\partial \frac{\partial}{\partial z}}{\partial \phi} \right).
$$

Performing some algebraic manipulation and after simplification, we finally get what we were looking for

$$
\Delta = \frac{\partial^{2}}{\partial \rho^{2}} + \frac{2}{\rho}\frac{\partial}{\partial \rho} + \frac{1}{\rho^{2}\sin^{2}\phi}\frac{\partial^{2}}{\partial \theta^{2}} + \frac{1}{\rho^{2}}\frac{\partial^{2}}{\partial \phi^{2}} + \frac{\cos \phi}{\rho^{2} \sin \phi} \frac{\partial}{\partial \phi}
$$

4) Using polar coordinates, starting with $\hat{{\bf i}}$ and $\hat{{\bf j}}$, we can define the associated polar vectors

\begin{align}
\hat{{\bf e}}_{r} = & \cos(\theta)\hat{{\bf i}} + \sin(\theta)\hat{{\bf j}}\\
\hat{{\bf e}}_{\theta} = & -\sin(\theta)\hat{{\bf i}} + \cos(\theta)\hat{{\bf j}}
\end{align}

Show that for a parametrized curve 

$$
{\bf r}(t) = x(t)\hat{{\bf i}} + y(t) \hat{{\bf j}}
$$

using $x(t) = r(t)\cos(\theta(t))$ and $y(t) = r(t)\sin(\theta(t))$, that

\begin{align}
{\bf r}(t) = & r(t) \hat{{\bf e}}_{r}\\
{\bf v}(t) = & \dot{r}\hat{{\bf e}}_{r} + r\dot{\theta}\hat{{\bf e}}_{\theta}\\
{\bf a}(t) = & \left(\ddot{r}-r\dot{\theta}^{2}\right)\hat{{\bf e}}_{r} + \left(2\dot{r}\dot{\theta}+r\ddot{\theta}\right)\hat{{\bf e}}_{\theta}
\end{align}

where ${\bf v} = \dot{{\bf r}}$ and ${\bf a} = \dot{{\bf v}}$.  Using this show that if the angular component of the acceleration is zero, i.e. ${\bf a}$ depends only on $\hat{{\bf e}}_{r}$, then the motion is constrained by the identity

$$
\dot{\theta}r^{2} = c
$$

where $c$ is a constant.  Note, this is Kepler's Second Law from orbital mechanics. 

**Solution**

From $x(t) = r(t)\cos(\theta(t))$ and $y(t) = r(t)\sin(\theta(t))$, we can re-write ${\bf r(t)}$ in polar coordinates

\begin{align}
{\bf r}(t) = r(t)\cos(\theta(t))\hat{{\bf i}} + r(t)\sin(\theta(t))\hat{\bf j}\\
= r(t)(\cos(\theta(t))\hat{{\bf i}}+\sin(\theta(t))\hat{\bf j})\\
= r(t)\hat{\bf e}_{r}
\end{align}

Now, since ${\bf v}(t)$ is ${\bf \dot r}(t)$, then

\begin{align}
{\bf v(}t) = \frac{\partial}{\partial r}(r(t)\cos(\theta(t))\hat{{\bf i}}+\frac{\partial}{\partial \theta}(r(t)\cos(\theta(t)))\hat{{\bf i}}+\frac{\partial}{\partial r}(r(t)\sin(\theta(t)))\hat{{\bf j}}+\frac{\partial}{\partial \theta}(r(t)\sin(\theta(t)))\hat{{\bf j}}\\
= \dot r(t)\cos(\theta(t))\hat{{\bf i}}-r(t)\dot\theta(t)\sin(\theta(t))\hat{{\bf i}}+\dot r(t)\sin(\theta(t))\hat{{\bf j}}+r(t)\dot\theta(t)\cos(\theta(t))\hat{{\bf j}}\\
= \dot r(t)(\cos(\theta(t))\hat{{\bf i}}+\sin(\theta(t))\hat{{\bf j}})+r(t)\dot\theta(t)(-\sin(\theta(t))\hat{\bf{i}}+\cos(\theta(t))\hat{\bf{j}})
\end{align}

For simplicity in the notation, we will drop the $t$ and rewrite the above equation in terms of the polar vectors to obtain

$$
{\bf v}(t)=\dot r\hat{\bf{e}}_{r}+r\dot \theta\hat{\bf{e}}_{\theta}
$$

Now for the acceleration ${\bf{a}}(t)= {\bf\dot v}(t)$ we derive

\begin{align}
{\bf\dot v}(t) = \ddot{r}\cos(\theta(t))\hat{{\bf i}}-\dot r\dot\theta\sin(\theta(t))\hat{{\bf i}}-\dot r\dot\theta\sin(\theta(t))\hat{{\bf i}}-r\dot\theta^{2}\cos(\theta(t))\hat{{\bf i}}-r\ddot\theta\sin(\theta(t))\hat{{\bf i}}\\
+\ddot{r}\sin(\theta(t))\hat{{\bf j}}+\dot r\theta\cos(\theta(t))\hat{{\bf j}}+\dot r\dot\theta\cos(\theta(t))\hat{{\bf j}}-r\dot\theta^{2}\sin(\theta(t))\hat{{\bf j}}+r\ddot\theta\cos(\theta(t))\hat{{\bf j}}
\end{align}

After some algebra and rearrengement, we get

\begin{align}
{\bf a}(t) = \ddot{r}(\cos(\theta(t))\hat{{\bf i}}+\sin(\theta(t))\hat{{\bf j}})+2\dot r\dot\theta(-\sin(\theta(t)\hat{{\bf i}}+\cos(\theta(t))\hat{{\bf j}})-r\dot\theta^{2}(\cos(\theta(t))\hat{{\bf i}}+\sin(\theta(t))\hat{{\bf j}})+r\ddot{\theta}(-\sin(\theta(t))\hat{{\bf i}}+\cos(\theta(t))\hat{{\bf j}}),
\end{align}

which can be rewritten as

$$
{\bf a}(t)=(\ddot{r}-r\dot\theta^{2})\hat{{\bf e}}_{r} + (2\dot r\dot\theta+r\ddot{\theta})\hat{{\bf e}}_{\theta}
$$

If the angular component of the acceleration is zero, that implies

$$
2\dot r\dot\theta+r\ddot{\theta}=0
$$

So we can write

$$
r(2\dot r\dot\theta+r\ddot{\theta})=0\longrightarrow 2r\dot r\dot\theta+r^{2}\ddot{\theta}=0
$$

We notice that the last espression is simply the derivative with respect to time of $\dot\theta r^{2}$. Therefore,

$$
\frac{d}{dt}(\dot\theta r^{2})=0
$$

which implies

$$
\dot\theta r^{2} = c,
$$

where $c$ is just a constant value $\in$ $\mathbb{R}$.

5) Using the previous problem, show that for the vector field 

$$
{\bf F} = M\hat{{\bf i}} + N\hat{{\bf j}} + P\hat{{\bf k}}
$$

in cylindrical coordinates i.e.

$$
x=r\cos(\theta), ~ y=r\sin(\theta), ~ z = z
$$

is given by

$$
{\bf F} = \tilde{M}\hat{{\bf e}}_{r} + \tilde{N}\hat{{\bf e}}_{\theta} + P\hat{{\bf k}}
$$

where

$$
\tilde{M} = \cos(\theta)M + \sin(\theta)N, ~ \tilde{N} = -\sin(\theta)M + \cos(\theta)N.
$$

Using this, show that the divergence 

$$
\nabla \cdot {\bf F} = M_{x} + N_{y} + P_{z}
$$

becomes

$$
\nabla \cdot {\bf F} = \frac{1}{r}\partial_{r}\left(r\tilde{M}\right) + \frac{1}{r}\tilde{N}_{\theta} + P_{z}
$$

**Solution**

Since in the conversion from cartesian coordinates to cylindrical coordinates nothing happens on $z$, then, for simplicity in the notation, we put it aside for a second and we will focus on the $x$ and $y$ components.

\begin{align}
{\bf F} = ({\bf F} \cdot \hat{{\bf e}}_{r})\hat{{\bf e}}_{r} + ({\bf F} \cdot \hat{{\bf e}}_{\theta})\hat{{\bf e}}_{\theta} \\
= (\cos(\theta)M + \sin(\theta)N)\hat{{\bf e}}_{r} + (-\sin(\theta)M + \cos(\theta)N) \hat{{\bf e}}_{\theta} \\
= \tilde{M}\hat{{\bf e}}_{r} + \tilde{N}\hat{{\bf e}}_{\theta}.
\end{align}

Now, we add the $z$ term, that has not changed in the transformation and we get: ${\bf F} = \tilde{M}\hat{{\bf e}}_{r} + \tilde{N}\hat{{\bf e}}_{\theta} + P\hat{{\bf k}}.$

In class,we already show that the gradient in cylindrical coordinates, once again putting, temporarily, aside $z$ is 

$$
\nabla = \hat{{\bf e}}_{r} \partial_{r} + \hat{{\bf e}}_{\theta}\left(\frac{1}{r} \partial_{\theta}\right);
$$

from here we can calculate the divergence:

$$
\nabla\cdot{\bf F} = \langle \partial_{r}, \frac{1}{r} \partial_{\theta} \rangle\cdot \langle \tilde{M}, \tilde{N} \rangle = \tilde{M} \partial_{r} + \frac{1}{r}\tilde{N}\partial_{\theta} = \frac{1}{r}\partial_{r}\left(r\tilde{M}\right) + \frac{1}{r}\tilde{N}_{\theta}
$$

Now we plug back the unchanged $z$ component and we get the desired result

$$
 \nabla\cdot{\bf F} = \frac{1}{r}\partial_{r}\left(r\tilde{M}\right) + \frac{1}{r}\tilde{N}_{\theta} + P_{z}.
$$
