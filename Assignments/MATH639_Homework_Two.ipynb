1) A two dimensional steady flow has velocity components

$$
u_{1} = x_{2}, ~ u_{2} = x_{1}.
$$

Show that the stream (and in this case path) lines are hyperbolas of the form 

$$
x_{1}^{2} - x_{2}^{2} = C,
$$

where $C$ is a constant.  Plot the flow pattern (see code below for starters) and convince yourself that it represents an irrotational flow (i.e. zero vorticity) in a right-angled corner.  

Hint: So, the above choice of velocity field implies the flow is described by the following differential equations

$$
\frac{dx_{1}}{dt} = x_{2}, ~ \frac{dx_{2}}{dt} = x_{1}.
$$

If you take Calculus out behind a wood-shed and beat it with a stick, you can then write

$$
\frac{dx_{1}}{dx_{2}} = \frac{x_{2}}{x_{1}}.
$$

Note, suppose you have a function of two variables, say $f(x_{1},x_{2})$.  Then if we look along level sets, i.e. those points $(x_{1},x_{2})$ such that

$$
f(x_{1},x_{2}) = C,
$$

then

$$
df = \partial_{x_{1}}f dx_{1} + \partial_{x_{2}}f dx_{2} = 0.
$$

I leave the rest up to you.

**Solution for Problem 1**

We are given $u_{1}=x_{2}$ and $u_{2}=x_{1}$. Following the calculus-beating hint, we have

$$
\frac{dx_{1}}{dx_{2}}=\frac{x_{2}}{x_{1}},
$$

which we can separate for the respective variables and integrate

$$
\int x_{1}dx_{1} = \int x_{2}dx_{2} \longrightarrow \frac{x_{1}^{2}}{2} + C_{1} = \frac{x_{2}^{2}}{2} + C_{2}.
$$

Multiplying both sides by $2$ and lumping the resulting constants together on the right-hand side and the variables on the left hand-side yields

$$
x_{1}^{2} -x_{2}^{2} = C
$$

```python
delta = 0.025
x = np.arange(-1.0, 1.0, delta) # To understand the right angle bit, you made need to adjust this
y = np.arange(-1.0, 1.0, delta) # To understand the right angle bit, you made need to adjust this
X, Y = np.meshgrid(x, y)
Z = (X**2.-Y**2.) # Yes, you need to figure this out.
plt.contour(X,Y,Z);
```

We know that the gradient is orthogonal to the the streamlines, in the direction of steepest ascent. Therefore, in a 45-degree-rotated frame of reference, we see that our velocity field is effectively given as the gradient of a scalar function, thus, the flow is irrotational.

2)  Consider a steady, axisymmetric flow of an incompressible fluid. In cylindrical coordinates $(r,\theta,z)$ this gives us the equation

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(ru_{r}\right) + \frac{\partial}{\partial z}\left(u_{z}\right) = 0
$$

Show how to define a streamfunction $\psi$ so that this equation is automatically satisfied.  

**Solution for Problem 2**


We have

$$
\frac{1}{r}\frac{\partial}{\partial r}\left(ru_{r}\right) + \frac{\partial}{\partial z}\left(u_{z}\right) = 0.
$$

Let pull a streamfuntion out of a hat

$$
\Psi(r,z,t).
$$

Let us define it so that

\begin{align}
u_{r}=\frac{1}{r}\partial_{z} \Psi\\
u_{z}=-\frac{1}{r}\partial_{r} \Psi
\end{align}

Let us plug our $u_{r}$ and $u_{z}$ in our original equation for a steady, axisymmetric flow of an incompressible fluid

$$
\frac{1}{r}\partial_{r}\left(r\frac{1}{r}\partial_{z} \Psi\right) + \partial_{z}\left(-\frac{1}{r}\partial_{r} \Psi\right) = 0
$$

which yields

$$
\frac{1}{r}\partial_{r}\partial_{z} \Psi - \frac{1}{r}\partial_{z}\partial_{r} \Psi = 0,
$$

since $\partial_{r}\partial_{z}=\partial_{z}\partial_{r}$, the equation is immediately satisfied.

3) In a plane Couette flow, we imagine a fluid moving between two plates.  The top plate, at $x_{2}=b$ is moving with horizontal velocity $U$ while the bottom plate at $x_{2}=0$ is stationary.  

![couette](https://upload.wikimedia.org/wikipedia/commons/9/93/Laminar_shear.svg)

At steady state, this gives us the velocity field

$$
u_{1} = \frac{Ux_{2}}{b}, ~ u_{2} = u_{3} = 0.
$$

Find the symmetric and anti-symmetric parts of the strain-rate tensor.  Along which directions do normal stresses appear?  What is the vorticity for this flow?  What is the circulation around a closed circle between the plates?  Show that the streamfunction is given by

$$
\psi(x_{1},x_{2}) = \frac{Ux_{2}^{2}}{2b} + C
$$

**Solution for Problem 3**

The strain-rate tensor is given by

$$
\overleftrightarrow{T_{ij}} = \overleftrightarrow{e_{ij}} +\frac{1}{2}\overleftrightarrow{r_{ij}},
$$

where

$$
\overleftrightarrow{e_{ij}} = \frac{1}{2}\left(\frac{\partial{u_{i}}}{\partial{x_{j}}} + \frac{\partial{u_{j}}}{\partial{x_{i}}}\right),
$$

is the symmetric component of $\overleftrightarrow{T_{ij}}$ and thus $\overleftrightarrow{e_{ij}}=\overleftrightarrow{e_{ji}}$, whereas

$$
\overleftrightarrow{r_{ij}} = \frac{\partial{u_{i}}}{\partial{x_{j}}} - \frac{\partial{u_{j}}}{\partial{x_{i}}} = -\overleftrightarrow{r_{ji}}
$$

is the antisymmetric component.

In our case, since we are working the $x_{1}x_{2}$-plane, we have

$$
\overleftrightarrow{e_{12}} = \frac{1}{2}\left(\frac{\partial{u_{1}}}{\partial{x_{2}}} + \frac{\partial{u_{2}}}{\partial{x_{1}}}\right)
$$

and

$$
\overleftrightarrow{r_{12}} = \frac{\partial{u_{1}}}{\partial{x_{2}}} - \frac{\partial{u_{2}}}{\partial{x_{1}}}
$$

So, since we are given that $u_{2}=0$, we only need to compute

$$
\frac{\partial{u_{1}}}{\partial{x_{2}}} = \frac{U}{b}.
$$

Thus, for the $1,2$ element of our symmetric tensor, we have

$$
e_{12}=\frac{U}{2b},
$$

so our tensor is given by
\begin{gather*}
\overleftrightarrow{e_{12}} =
\begin{bmatrix}
0 & \frac{U}{2b}\\
\frac{U}{2b} & 0
\end{bmatrix}
\end{gather*}

whereas, for the antisymmetric tensor we have

\begin{gather*}
\overleftrightarrow{r_{12}} = 
\begin{bmatrix}
0 & \frac{U}{2b}\\
-\frac{U}{2b} & 0
\end{bmatrix}
\end{gather*}

Now on to the vorticity. By definition, it is given by

$$
\textbf{$\omega$} = \nabla \times \textbf{u},
$$

which can be readily computed as the determinant of the following matrix

\begin{bmatrix}
\hat{{\bf i}} & \hat{{\bf j}} & \hat{{\bf k}}\\
\partial_{x_{1}} & \partial_{x_{2}} & \partial_{x_{3}}\\
u_{1} & u_{2} & u_{3}
\end{bmatrix} 

i.e. 

\begin{gather*}
\textbf{$\omega$} =
\begin{bmatrix}
\hat{{\bf i}} & \hat{{\bf j}} & \hat{{\bf k}}\\
\partial_{x_{1}} & \partial_{x_{2}} & \partial_{x_{3}}\\
\frac{Ux_{2}}{b} & 0 & 0,
\end{bmatrix}
\end{gather*}

which yields

$$
\textbf{$\omega$} = -\frac{U}{b}\hat{{\bf k}}
$$

Now we have to find the circulation around a close circle between the plates. Thus

\begin{align}
\Gamma = \oint_{C} {\bf u}\cdot d{\bf s}\\
=\oint_{A} (\nabla\times u)_{3}dA\\
=\oint_{A}\omega_{3}dA,
\end{align}

where $\omega_{3}=-\frac{U}{b}$. Then, given the geometry of the problem, we switch to polar coordinates. As the radius of the circle is constant, we solve for

\begin{align}
\Gamma = -\int_{0}^{2\pi}\frac{Ur}{b}rd\theta\\
= -\frac{U}{b}r^{2}\int_{0}^{2\pi}d\theta\\
= -\frac{2U}{b}\pi r^{2}
\end{align}

So, for a positive $U$ we have a negative circulation, while for a negative $U$ (i.e. moving in the $-x_{1}$ direction) we will have a positive circulation.

Now we are asked to show that

$$
\psi(x_{1},x_{2}) = \frac{Ux_{2}^{2}}{2b} + C
$$

is the streamfunction for our fluid. Let us define $\psi(x_{1},x_{2})$ such that

\begin{align}
u_{1} = \partial_{x_{2}}\psi\\
u_{2} = -\partial_{x_{1}}\psi
\end{align}

The vorticity is given by

$$
\omega = \partial_{x_{1}}u_{2} - \partial_{x_{2}}u_{1};
$$

since $\partial_{x_{1}}u_{2}=0$, using our streamfunction, we get

\begin{align}
\omega = - \partial_{x_{2}}\partial_{x_{2}}\psi\\
= -\partial^{2}_{x_{2}}\psi
\end{align}

Thus
$$
\partial^{2}_{x_{2}}\psi = \frac{U}{b}
$$

So, integrating that expression once with respect to $x_{2}$ yields

$$
\int\partial^{2}_{x_{2}}\psi dx_{2} = \int \frac{U}{b}dx_{2} \longrightarrow \partial_{x_{2}}\psi = \frac{Ux_{2}}{b}
$$

and then integrating a second time (and lumping all constants of integration together) yields

$$
\int \partial_{x_{2}}\psi dx_{2} = \int \frac{Ux_{2}}{b}dx_{2} \longrightarrow \psi = \frac{Ux_{2}^{2}}{2b} + C,
$$

which is our expected result.

4) Using Stoke's Theorem for a vector field ${\bf F}$

$$
\oint_{C} {\bf F} \cdot d{\bf s} = \iint_{S} \left(\nabla \times {\bf F}\right)\cdot \hat{{\bf n}} d\sigma
$$

where $C$ is a simple, closed contour, and $S$ is some surface whose boundary is $C$, and using the Divergence Theorem where, for _ closed _ surface $\tilde{S}$ we have 

$$
\iint_{\tilde{S}}{\bf F} \cdot \hat{{\bf n}} d\sigma = \iiint_{V} \nabla \cdot {\bf F} dV, 
$$

where $V$ is contained within $\tilde{S}$ that

$$
\nabla \cdot \left(\nabla \times {\bf F} \right) = 0.
$$

Note, by using the integral identities, you are showing this identity is true in any coordinate system.

**Solution for Problem 4**

So,for a closed contour $C$, according to Stoke's Theorem we have

$$
\oint_{C} {\bf F} \cdot d{\bf s} = \iint_{S} \left(\nabla \times {\bf F}\right)\cdot \hat{{\bf n}} d\sigma,
$$

where $S$ is some surface whose boundary is $C$. Thus, let

$$
\tilde{{\bf F}} = \nabla \times {\bf F}
$$

be a vector field. Then, by the means of the Divergence Theorem we have


$$
\iint_{\tilde{S}}{\bf \tilde{F}} \cdot \hat{{\bf n}} d\sigma = \iiint_{V} \nabla \cdot {\bf \tilde{F}} dV, 
$$

where $V$ is a volume contained within the surface $\tilde{S}$. So, the integrand on the right-hand side is

$$
\nabla \cdot {\bf \tilde{F}} = \nabla \cdot (\nabla \times {\bf F}).
$$

Let define ${\bf F}$ as

$$
{\bf F} = P\hat{\bf {i}} + Q\hat{\bf {j}} + R\hat{\bf {k}}.
$$

Then, computing its curl yields

$$
\nabla \times {\bf F} = \left< \frac{\partial{R}}{\partial{x_{2}}} - \frac{\partial{Q}}{\partial{x_{3}}}, \frac{\partial{P}}{\partial{x_{3}}} - \frac{\partial{R}}{\partial{x_{1}}}, \frac{\partial{Q}}{\partial{x_{1}}} - \frac{\partial{P}}{\partial{x_{2}}}\right>.
$$

Let us call this vector $\textbf{$\omega$}$.
Now, taking the divergence of this vector yeids

\begin{align}
\nabla \cdot \textbf{$\omega$} = \frac{\partial^{2}{R}}{\partial{x_{1}}\partial{x_{2}}} - \frac{\partial^{2}{Q}}{\partial{x_{1}}\partial{x_{3}}} + \frac{\partial^{2}{P}}{\partial{x_{2}}\partial{x_{3}}} - \frac{\partial^{2}{R}}{\partial{x_{2}}\partial{x_{1}}} + \frac{\partial^{2}{Q}}{\partial{x_{3}}\partial{x_{1}}} - \frac{\partial^{2}{P}}{\partial{x_{3}}\partial{x_{2}}}\\
= 0
\end{align}

since mixed partial derivatives are equal to each other and therefore all the terms above cancel out, by Clairaut's Theorem.

Therefore we have

$$
\nabla \cdot \left(\nabla \times {\bf F} \right) = 0,
$$

and thus 

$$
\iint_{\tilde{S}}{\bf \tilde{F}} \cdot \hat{{\bf n}} d\sigma = \iiint_{V} \nabla \cdot {\bf \tilde{F}} dV = 0, 
$$

from the integral identities, this shows this identity to be true in any coordinate system.

5) Suppose we have an unsteady flow given via the velocity field

$$
u_{1}(x_{1},x_{2},t) = \frac{x_{1}}{1+t}, ~ u_{2}(x_{1},x_{2},t) = \frac{2 x_{2}}{2+t}.
$$

First, plot different families of _ path lines _ in the $(x_{1},x_{2})$ plane.  This means solve the corresponding differential equations for initial conditions

$$
x_{1}(0) = x_{10}, ~ x_{2}(0) = x_{20},
$$

and then getting rid of the explicit parameter $t$.  Desribe the flows for different initial conditions by way of plotting using Python.  The standard command for plotting

`plt.plot(x,y)`

should suffice.  

**Solution for Problem 5**

We have $u_{1}=\dot x_{1}$ and $u_{2}=\dot x_{2}$, therefore

$$
\frac{dx_{1}}{dt} = \frac{x_{1}}{1+t}
$$

and

$$
\frac{dx_{2}}{dt} = \frac{2 x_{2}}{2+t}.
$$

Let us look at the first integral. Separating the variables and integrating yields

$$
x_{1} = C_{1}(1+t);
$$

we are given $x_{1}(0) = x_{10}$, therefore $C_{1}= x_{10}$ and our solution for given initial data is

$$
x_{1}(t)= x_{10}(1+t).
$$

Now, for the second integral we get

$$
x_{2} = C_{2}(2+t)^{2};
$$

for the given initial condition $x_{2}(0) = x_{20}$, we get $C_{2}=\frac{x_{20}}{4}$ and thus our solution is given by

$$
x_{2}(t) = x_{20}\frac{(2+t)^{2}}{4}.
$$

Now, to get rid of the explicit parameter $t$, we solve $x_{1}$ for t, which yields

$$
t = \frac{x_{1}-x_{10}}{x_{10}}.
$$

Then, we plug that into our formula for $x_{2}(t)$, which yields

$$
x_{2} = x_{20}\frac{\left(2+\frac{x_{1}-x_{10}}{x_{10}}\right)^{2}}{4}.
$$

Therefore we see that our path lines are parabolas, whose concavity will ultimately depend on the sign of $x_{20}$. We can plot our results for different values of $x_{10}$ and $x_{20}$

```python
x10 = 1
x20 = 1
x1vals = np.linspace(-5,5,int(1e2))
x2vals = (x20/4.)*(2 + (x1vals-x10)/x10)**2

plt.plot(x1vals,x2vals);
plt.xlabel('$x_{1}$');
plt.ylabel('$x_{2}$');
```

```python
x10 = .1
x20 = -.1
x1vals = np.linspace(-5,5,int(1e2))
x2vals = (x20/4.)*(2 + (x1vals-x10)/x10)**2

plt.plot(x1vals,x2vals);
plt.xlabel('$x_{1}$');
plt.ylabel('$x_{2}$');
```
