Problem 1) Show for a general vector field ${\bf u}$ that 

$$
{\bf u}\cdot \nabla {\bf u} = \frac{1}{2}\nabla \left|{\bf u}\right|^{2} + {\bf \omega}\times {\bf u}
$$

where the vorticity $\omega = \nabla \times {\bf u}$.

**Solution to Problem 1**

Looking at the right side of the equality:
$$
\left|{\bf u}\right|^{2} = u_{1}^{2} + u_{2}^{2} + u_{3}^{2} \\
$$

therefore,
$$
\frac{1}{2}\nabla \left|{\bf u}\right|^{2} = \frac{1}{2} \left< \partial x_{1} \left(u_{1}^{2} + u_{2}^{2} + u_{3}^{2}\right), \partial x_{2} \left(u_{1}^{2} + u_{2}^{2} + u_{3}^{2}\right), \partial x_{3} \left(u_{1}^{2} + u_{2}^{2} + u_{3}^{2}\right) \right>.
$$

We now look at ${\bf \omega}\times {\bf u} $:

$$
{\bf \omega}\times {\bf u} = \left|\begin{matrix}
\hat{{\bf i}} & \hat{{\bf j}} & \hat{{\bf k}}\\
\omega_{1} & \omega_{2} & \omega_{3}\\
u_{1} & u_{2} & u_{3}
\end{matrix}\right| = \left<u_{3}\omega_{2} - u_{2}\omega_{3}, u_{1}\omega_{3} - u_{3}\omega_{1}, u_{2}\omega_{1} - u_{1}\omega_{2} \right> \\
\\
= \left<(u_{3}\partial x_{3} u_{1} - u_{3}\partial x_{1} u_{3}) - (u_{2}\partial x_{1} u_{2} - u_{2}\partial x_{2} u_{1}), (u_{3}\partial x_{3} u_{2} - u_{3}\partial x_{2} u_{3}) + (u_{1}\partial x_{1} u_{2} - u_{1}\partial x_{2} u_{1}), (u_{2}\partial x_{2} u_{3} - u_{2}\partial x_{3} u_{2}) - (u_{1}\partial x_{3} u_{1} - u_{1}\partial x_{1} u_{3})\right> \\
\\
= \left<(u_{3}\partial x_{3} u_{1} + u_{2}\partial x_{2} u_{1} - \frac{1}{2}\partial x_{1}(u_{3}^{2}+u_{2}^{2})), (u_{3}\partial x_{3} u_{2} + u_{1}\partial x_{1} u_{2} - \frac{1}{2}\partial x_{2}(u_{3}^{2}+u_{1}^{2})), (u_{2}\partial x_{2} u_{3} + u_{1}\partial x_{1} u_{3} - \frac{1}{2}\partial x_{1}(u_{2}^{2}+u_{1}^{2})) \right> 
$$

So, the right side of the equation can be written as 

$$
\frac{1}{2}\nabla \left|{\bf u}\right|^{2} + {\bf \omega}\times {\bf u} = \\
\left<(u_{3}\partial x_{3} u_{1} + u_{2}\partial x_{2} u_{1} + \frac{1}{2}\partial x_{1}u_{1}^{2}), (u_{3}\partial x_{3} u_{2} + u_{1}\partial x_{1} u_{2} + \frac{1}{2}\partial x_{2}u_{2}^{2}), (u_{2}\partial x_{2} u_{3} + u_{1}\partial x_{1} u_{3} + \frac{1}{2}\partial x_{3}u_{3}^{2}) \right> \\
= \left<(u_{3}\partial x_{3} u_{1} + u_{2}\partial x_{2} u_{1} + u_{1}\partial x_{1} u_{1}), (u_{3}\partial x_{3} u_{2} + u_{1}\partial x_{1} u_{2} + u_{2}\partial x_{2} u_{2}), (u_{2}\partial x_{2} u_{3} + u_{1}\partial x_{1} u_{3} + u_{3}\partial x_{3} u_{3}) \right> \\
= \left<(u_{1}\partial x_{3} u_{3} + u_{1}\partial x_{2} u_{2} + u_{1}\partial x_{1} u_{1}), (u_{2}\partial x_{3} u_{3} + u_{2}\partial x_{1} u_{1} + u_{2}\partial x_{2} u_{2}), (u_{3}\partial x_{2} u_{2} + u_{3}\partial x_{1} u_{1} + u_{3}\partial x_{3} u_{3}) \right> \\
= \left<(u_{1}(\partial x_{3} u_{3} + \partial x_{2} u_{2} + \partial x_{1} u_{1})), (u_{2}(\partial x_{3} u_{3} + \partial x_{1} u_{1} + \partial x_{2} u_{2})), (u_{3}(\partial x_{2} u_{2} + \partial x_{1} u_{1} + \partial x_{3} u_{3})) \right> \\
=(\partial x_{1} u_{1} + \partial x_{2} u_{2} + \partial x_{3} u_{3}) \left<u_{1}, u_{2}, u_{3}\right>  \\
= {\bf u}\cdot \nabla {\bf u}
$$

Therefore we see that the two sides of the equations match.

Problem 2) Show for a planar incompressible and inviscid flow, i.e. one restricted to the $(x,y)$ plane for example, that 

$$
\frac{D\omega}{Dt} = 0,
$$

where $\omega$ is the scalar magnitude of the vorticity and $D/Dt$ is the material derivative.  What does this tell you about voriticy in planar flows?  

**Solution to Problem 2**

The material derivative is given by 

$$
\frac{D}{Dt} = {\bf u} \cdot \nabla + \frac{\partial}{\partial t}
$$

and Bernoulli's equation for an incompressible and inviscid flow

$$
{\bf u}_{t} + {\bf u}\cdot \nabla {\bf u} = -\frac{1}{\rho_{0}}\nabla \overline{p} - {\bf g}.
$$


Therefore we can write

$$
\frac{D{\bf u}}{Dt} = -\frac{1}{\rho_{0}}\nabla \overline{p} - {\bf g} \\
\Rightarrow \nabla \times \frac{D{\bf u}}{Dt} =\nabla \times \left(-\frac{1}{\rho_{0}}\nabla\overline{p} - {\bf g}\right) \\
\Rightarrow \frac{D}{Dt} (\nabla \times {\bf u}) = 0 \\
\Rightarrow {\bf \hat{e}_{3}} \frac{D}{Dt} (\nabla \times {\bf u}) = 0 \\
\Rightarrow \frac{D}{Dt} (\nabla \times {\bf u}){\bf \hat{e}_{3}} = 0, \\
$$


where
\begin{gather*}
{\bf \hat{e}_{3}} =
\begin{bmatrix}
0\\
0\\
1
\end{bmatrix}
\end{gather*}

Therefore, since for a flow in the $x_{1}x_{2}$-plane we have 

$$
(\nabla \times {\bf u}){\bf \hat{e}_{3}} = (\partial_{x_{1}}u_{2}-\partial_{x_{2}}u_{1}) = \omega,
$$ 

then

$$
\frac{D\omega}{Dt} = 0,
$$

for a scalar vorticity.

Therefore we see that the sum of the Eulerian and Lagrangian component of the vorticity is zero, thus the vorticity for a planar incompressible and inviscid flow is constant along the flow and at any point of it. 

Problem 3) Following in the vein of the doublet, if we define the complex function $f(z;\epsilon)$ where 

$$
f(z;\epsilon) = \frac{m}{2\pi}\left(\ln\left(z+3\epsilon\right) - \ln\left(z+2\epsilon\right) + \ln\left(z-3\epsilon\right) - \ln\left(z-2\epsilon\right) \right),
$$

and then let the source/sink strengths $m$ go to infinity as $\epsilon$ goes to zero so that 

$$
\lim_{\epsilon \rightarrow 0} \frac{5m\epsilon^{2}}{2\pi} = \mu,
$$

then describe the flow given by the complex function

$$
f_{q}(z) = \lim_{\epsilon\rightarrow 0}f(z;\epsilon).
$$

Provide plots and, using Bernoulli's equation with density normalized to being one and ignoring gravity, a description of the pressure field associated with this plow.  

**Solution to Problem 3**

Using some basic properties of logarithms and doing some algebra, we rewrite our function as:

$$
f(z;\epsilon) = \frac{m}{2\pi}\left(\ln(z^{2}-9\epsilon^{2})-\ln(z^{2}-4\epsilon^{2})\right) \\
= \frac{m}{2\pi}\ln \left(\frac{z^{2}-9\epsilon^{2}}{z^{2}-4\epsilon^{2}}\right)\\
= \frac{m}{2\pi}\ln \left(\frac{1-9\frac{\epsilon^{2}}{z^{2}}}{1-4\frac{\epsilon^{2}}{z^{2}}}\right)
$$

Using a geometric series argument:

$$
= \frac{m}{2\pi}\ln \left(\left(1-9\frac{\epsilon^{2}}{z^{2}}\right)\left(1+4\frac{\epsilon^{2}}{z^{2}} + \mathcal{O}(\epsilon^{4})\right)\right) \\
= \frac{m}{2\pi}\ln \left(1-9\frac{\epsilon^{2}}{z^{2}}+4\frac{\epsilon^{2}}{z^{2}} + \mathcal{O}(\epsilon^{4})\right) \\
= \frac{m}{2\pi}\ln \left(1-5\frac{\epsilon^{2}}{z^{2}}+ \mathcal{O}(\epsilon^{4})\right) \\
= \frac{m}{2\pi}\left(5\frac{\epsilon^{2}}{z^{2}}+ \mathcal{O}(\epsilon^{4})\right) \\
= \frac{5m\epsilon^{2}}{2\pi z^{2}} + \mathcal{O}(\epsilon^{4})
$$

Since $\lim_{\epsilon \rightarrow 0}\frac{5m\epsilon^{2}}{2\pi} = \mu,$ 

$$
f_{q}(z) = \frac{\mu}{z^{2}} \\
= \frac{\mu}{r^{2}e^{2i\theta}} = \frac{\mu}{r^{2}}e^{-2i\theta} \\
= \frac{\mu}{r^{2}}(\cos(2\theta) - i\sin(2\theta))
$$

Thus, our streamfunctions turn out to be

$$
\phi = \frac{\mu}{r^{2}}\cos(2\theta); ~ ~ ~ \psi = -\frac{\mu}{r^{2}}\sin(2\theta).
$$

$$
f_{q}'(z) = -\frac{2\mu}{z^{3}} = -\frac{2\mu}{r^{3}}e^{-3i\theta} \longrightarrow f_{q}'(z) = -\frac{2\mu}{r^{3}}(\cos(3\theta) - i\sin(3\theta))
$$

which gives us, 

$$
u = -\frac{2\mu}{r^{3}}\cos(3\theta);
$$

$$
v = -\frac{2\mu}{r^{3}}\sin(3\theta)
$$

A plot of some level set solutions to the stream function $\psi$ is shown below.

We use Bernoulli's Equation:

$$
\frac{\rho_{0}}{2}\left|f_{q}'(z)\right|^{2} + \overline{p} = \overline{p}_{\infty} \\
\Rightarrow \frac{\rho_{0}}{2}\left|-\frac{2\mu}{r^{3}}(\cos(3\theta) - i\sin(3\theta))\right|^{2} + \overline{p} = \overline{p}_{\infty} \\
\Rightarrow \frac{\rho_{0}}{2}\left(\left(-\frac{2\mu}{r^{3}}\right)^{2}\cos^{2}(3\theta) + \left(-\frac{2\mu}{r^{3}}\right)^{2}\sin^{2}(3\theta))\right) + \overline{p} = \overline{p}_{\infty} \\
\Rightarrow \frac{\rho_{0}}{2} \left(\frac{4\mu^{2}}{r^{6}}\right) + \overline{p} = \overline{p}_{\infty}
$$

Since density has been normalized to 1,

$$
\overline{p} = \overline{p}_{\infty} - \left(\frac{2\mu^{2}}{r^{6}}\right),
$$

or

$$
\overline{p}_{\infty}-\overline{p}=\left(\frac{2\mu^{2}}{r^{6}}\right),
$$

so we see that the pressure field is inversely proportional to $r^{6}$, which means that as $r\rightarrow\infty$, we have that $\overline{p}=\overline{p}_{\infty}$.

Now we plot our streamfunction $\psi$

```python
def velocity_field(psi):
    u = sympy.lambdify((x, y), psi.diff(y), 'numpy')
    v = sympy.lambdify((x, y), -psi.diff(x), 'numpy')
    return u, v

def plot_streamlines(ax, u, v, xlim=(-1, 1), ylim=(-1, 1)):
    x0, x1 = xlim
    y0, y1 = ylim
    Y, X =  np.ogrid[y0:y1:100j, x0:x1:100j]
    ax.streamplot(X, Y, u(X, Y), v(X, Y), color='blue')
        
def format_axes(ax):
    ax.set_aspect('equal')
    ax.figure.subplots_adjust(bottom=0, top=1, left=0, right=1)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    for spine in ax.spines.itervalues():
        spine.set_visible(False)
```

```python
pi = np.pi
m=1
def doubletlike_stream_function(n=0, U=0):
    r = sympy.sqrt(x**2 + y**2)
    theta = sympy.atan2(y, x)
    return U - (m/r)*sympy.sin(2*theta) #psi
```

```python
import matplotlib.pyplot as plt

psi = doubletlike_stream_function()
u, v = velocity_field(psi)

xlim = ylim = (-3, 3)
fig, ax = plt.subplots(figsize=(4, 4))
plot_streamlines(ax, u, v, xlim, ylim)

c = plt.Circle((0, 0), radius=1, facecolor='none')
ax.add_patch(c)

format_axes(ax)
```

Our streamlines make up a quadrupole "spreading out" from the singularity.

Problem 4) The following function $f_{R}(z)$ where 

$$
f_{R}(z) = U_{\infty} z + \frac{m}{2\pi}\ln z,
$$

is known as the _Ranking Leading Edge_.  Describe the flow by providing stream-function plots and qualitative descriptions of said plots.  A _stagnation point_ is a point in the flow where the velocity is zero.  Where are the stagnation points in the Rankine Leading Edge?  Using Bernoulli's equation with the gravity set to zero, can you describe the pressure distribution in this flow?  

**Solution to Problem 4**

Let $z=re^{i\theta}$, then

\begin{align}
f_{R}(z)=U_{\infty}re^{i\theta}+\frac{m}{2\pi}\ln(re^{i\theta})\\
= U_{\infty}re^{i\theta}+\frac{m}{2\pi}(\ln(r)+i\theta)\\
= U_{\infty}r(\cos(\theta)+i\sin(\theta))+\frac{m}{2\pi}(\ln(r)+i\theta)
\end{align}

Where we used $e^{i\theta}=\cos(\theta)+i\sin(\theta)$.

Factoring real an imaginary parts yields 

$$
f_{R}(z)= \left(U_{\infty}r\cos(\theta)+\frac{m}{2\pi}\ln(r)\right) + i\left(U_{\infty}r\sin(\theta)+\frac{m\theta}{2\pi}\right),
$$

therefore our streamfunctions are

\begin{align}
\phi=U_{\infty}r\cos(\theta)+\frac{m}{2\pi}\ln(r),\\
\psi=U_{\infty}r\sin(\theta)+\frac{m\theta}{2\pi},
\end{align}

which we could also rewrite in Cartesian Coordinates

\begin{align}
\phi=U_{\infty}x+\frac{m}{4\pi}\ln(x^{2}+y^{2}),\\
\psi=U_{\infty}y+\frac{m}{2\pi}\tan^{-1}\left(\frac{y}{x}\right),
\end{align}

to help build some intuition about the flow. That being said, polar coordinates makes the algebra quite easier so we will continue to use them to derive the next results.

We have

$$
f^{\prime}(z)=U_{\infty}+\frac{m}{2\pi z}.
$$

Letting $z=re^{i\theta}$ yields

\begin{align}
f^{\prime}(z)=U_{\infty}+\frac{m}{2\pi r}e^{-i\theta}\\
= U_{\infty}+\frac{m}{2\pi r}(\cos(\theta)-i\sin(\theta)).
\end{align}

Then, in accordance with the Cauchy-Riemann equations we can find the velocity components

$$
u = U_{\infty}+\frac{m}{2\pi r}{\cos(\theta)},\\
$$

$$
v = \frac{m}{2\pi r}\sin(\theta).
$$

So to find the stagnation points we need to solve $u(r,\theta)=v(r,\theta)=0$, therefore

$$
v=0 \longrightarrow \frac{m}{2\pi r}\sin(\theta) = 0 \longrightarrow \theta = 0, ~ n\pi
$$

where $n\in\mathbb{Z}$.

Plugging this result back into $u$ yields

$$
r=-\frac{mU_{\infty}}{2\pi}
$$

for $n=0$ or even, and

$$
r=\frac{mU_{\infty}}{2\pi}
$$

for $n$ odd.

So, our stagnation points are

$$
(r_{\ast},\theta_{\ast})=\left(-\frac{mU_{\infty}}{2\pi},n\pi\right)
$$

for $n=2k$, with $k\in\mathbb{Z_{0}}$, and

$$
(r_{\ast},\theta_{\ast})=\left(\frac{mU_{\infty}}{2\pi},n\pi\right)
$$

for $n=2k+1$, with $k\in\mathbb{Z_{0}}$.

Using Bernoulli Equation with $g=0$ we can describe the pressure distribution in the flow. Thus

$$
\frac{\rho}{2}|f^{\prime}(z)|^{2}+\overline{p}=\frac{\rho}{2}U_{\infty}^{2}+\overline{p}_{\infty},
$$

where 

\begin{align}
|f^{\prime}(z)|^{2}=\left(U_{\infty}+\frac{m^{2}}{2\pi r}e^{-i\theta}\right)^{2}\\
= \left(U_{\infty}+\frac{m}{2\pi r}\cos(\theta)\right)^{2}+\frac{m^{2}}{4\pi^{2}r^{2}}\sin^{2}(\theta)\\
= U_{\infty}^{2}+\frac{mU_{\infty}}{\pi r}\cos(\theta)+\frac{m^{2}}{4\pi^{2} r^{2}}\cos^{2}(\theta)+\frac{m^{2}}{4\pi^{2}r^{2}}\sin^{2}(\theta)\\
= U_{\infty}^{2}+\frac{mU_{\infty}}{\pi r}\cos(\theta)+\frac{m^{2}}{4\pi^{2} r^{2}}
\end{align}

So, we plug this back into Bernoulli's equation and, when the algebra and the dust settle, we get

$$
\overline{p}=\overline{p}_{\infty}-\frac{\rho}{2}\left(\frac{mU_{\infty}}{\pi r}\cos(\theta)+\frac{m^{2}}{4\pi^{2}r^{2}}\right)
$$

or

\begin{align}
\overline{p}_{\infty}-\overline{p}=\frac{\rho}{2}\left(\frac{mU_{\infty}}{\pi r}\cos(\theta)+\frac{m^{2}}{4\pi^{2}r^{2}}\right)\\
= \frac{m\rho}{2\pi r}\left(U_{\infty}\cos(\theta)+\frac{m}{4\pi r}\right)
\end{align}

We can now plot the streamlines

```python
def velocity_field(psi):
    u = sympy.lambdify((x, y), psi.diff(y), 'numpy')
    v = sympy.lambdify((x, y), -psi.diff(x), 'numpy')
    return u, v

def plot_streamlines(ax, u, v, xlim=(-1, 1), ylim=(-1, 1)):
    x0, x1 = xlim
    y0, y1 = ylim
    Y, X =  np.ogrid[y0:y1:100j, x0:x1:100j]
    ax.streamplot(X, Y, u(X, Y), v(X, Y), color='blue')
        
def format_axes(ax):
    ax.set_aspect('equal')
    ax.figure.subplots_adjust(bottom=0, top=1, left=0, right=1)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    for spine in ax.spines.itervalues():
        spine.set_visible(False)
```

```python
pi = np.pi
m=1
def psi_stream_function(n=1, U=1):
    r = sympy.sqrt(x**2 + y**2)
    theta = sympy.atan2(y, x)
    return U*r*sympy.sin(theta)+(m*theta)/(2*pi) #psi
```

```python
import matplotlib.pyplot as plt

psi = psi_stream_function()
u, v = velocity_field(psi)

xlim = ylim = (-3, 3)
fig, ax = plt.subplots(figsize=(4, 4))
plot_streamlines(ax, u, v, xlim, ylim)

c = plt.Circle((0, 0), radius=1, facecolor='none')
ax.add_patch(c)

format_axes(ax)
```

We see that the streamlines are for the most part parallel to the $x$-axis and they go around the singularity at the origin. We see that the flow is directed in the positive $x$-direction.

Problem 5) So following up on integrals over curves, we can readily integrate functions $f(z)$, $z=x+iy$, along contours $C$ in the complex plane $\mathbb{C}$.  This is done via parametrization in exactly the same way you find any other line integral so that 

$$
\int_{C} f(z) dz = \int_{t_{0}}^{t_{f}} f(z) \frac{dz}{dt} dt = \int_{t_{0}}^{t_{f}} \left(\phi(x(t),y(t)) + i\psi(x(t),y(t))\right)\left(\frac{dx}{dt} + i \frac{dy}{dt} \right) dt
$$

So, choosing your contour to be $C_{r}$, i.e. the circle of radius $r$ centered at the origin, find 

$$
\oint_{C_{r}} z^{n} dz, ~ n \in \mathbb{Z}^{+}.
$$

Hint: Use $z = re^{i\theta}$, and let $\theta$ be the parameter with which you parameterize the circle.  Based on this result, suppose you have a function $f(z)$ which is analytic around the origin, so that we can expand it in a Taylor series so that 

$$
f(z) = \sum_{j=0}^{\infty} \frac{1}{j!}\left.\frac{d^{j}f}{dz^{j}}\right|_{z=0} z^{j}.
$$

What is 

$$
\oint_{C_{r}} f(z) dz 
$$

What does this tell you about integrating analytic functions in closed loops?  

Finally, now find 

$$
\oint_{C_{r}} z^{n} dz, ~ n \in \mathbb{Z}^{-},
$$

i.e. compute the contour integrals for $n$ a negative integer.  There is only one value of $n$ which gives an interesting result.  What is it?  Note, these integrals are actually fine since you are integrating _around_ the singularity, not over it like you would in real-variables.  This kind of trickery gives rise to the general notion that anything that is a difficulty in real variables can be gotten around by going into the complex plane.

**Solution to Problem 5**

To solve

$$
\oint_{C_{r}} z^{n} dz, ~ n \in \mathbb{Z}^{+},
$$

we let $z=re^{i\theta}$, thus

$$
\frac{dz}{d\theta}=ire^{i\theta}
$$

so our line integral becomes

$$
\oint_{C_{r}}z^{n}\frac{dz}{d\theta}d\theta\longrightarrow \oint_{C_{r}}(re^{i\theta})^{n}(ire^{i\theta})d\theta\longrightarrow \oint_{C_{r}}ir^{n+1}e^{(n+1)i\theta}d\theta
$$

Since, we have paramatrized with respect to $\theta$ and we are evaluating the line integral of a circle, we have

$$
0\leq\theta\leq2\pi;
$$

now, we use Euler's formula and, to simplify our notation, let $n+1=\xi$. Then

$$
\int_{0}^{2\pi}r^{\xi}(\cos(\xi\theta)+i\sin(\xi\theta))d\theta,
$$

thus

\begin{align}
r^{\xi}\int_{0}^{2\pi}(\cos(\xi\theta)+ir^{\xi}\int_{0}^{2\pi}\sin(\xi\theta))d\theta\\
= \frac{r^{\xi}}{\xi}[\sin(\xi\theta)]_{0}^{2\pi} - i\frac{r^{\xi}}{\xi}[\cos(\xi\theta)]_{0}^{2\pi}\\
= \frac{r^{n+1}}{n+1}[\sin((n+1)\theta)]_{0}^{2\pi} - i\frac{r^{n+1}}{n+1}[\cos((n+1)\theta)]_{0}^{2\pi}\\
= 0
\end{align}

for any $n\in\mathbb{Z}^{+}$.

Now, for a function $f(z)$ that has singularity at the origin, but it's analytic around it so that we can Taylor-expand it so that

$$
f(z) = \sum_{j=0}^{\infty} \frac{1}{j!}\left.\frac{d^{j}f}{dz^{j}}\right|_{z=0} z^{j},
$$

then

$$
\oint_{C_{r}}=0,
$$

so we see that integration of analytic functions over closed loops is independent of path.

Now for $n\in\mathbb{Z}^{-}$, our integral

$$
\int_{0}^{2\pi}r^{\xi}(\cos(\xi\theta)+i\sin(\xi\theta))d\theta
$$

yields

$$
\frac{r^{n+1}}{n+1}[\sin((n+1)\theta)]_{0}^{2\pi} - i\frac{r^{n+1}}{n+1}[\cos((n+1)\theta)]_{0}^{2\pi}
$$

which we see is $0$ only for $n=-1$. For any other $n\in\mathbb{Z}^{-}$, it would be equal to $2i\frac{r^{n+1}}{n+1}$.
